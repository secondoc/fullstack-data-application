# Création de la boucle

```python
URLloop = 'https://en.wikipedia.org/wiki/Wikipedia:Top_25_Report/'
data = pd.DataFrame(columns=columns2)

for d in range (len(url_list)) :
    URL = URLloop + url_list[d]
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table', {'class' : 'wikitable'}).tbody
    rows = table.find_all('tr')
    for i in range(1, len(rows)):
        tds = rows[i].find_all('td')
        if len(tds)>6:
            values=[tds[0].text.replace('\n', ''), tds[1].text.replace('\n', ''), tds[2].find('img').get('alt'), 
                    tds[3].text.replace('\n', ''), tds[4].text.replace('\n', ''), tds[5].text.replace('\n', '')]
            data=data.append(pd.Series(values, index=columns), ignore_index=True)
        if len(tds)==4 or len(tds)==5:
            if len(rows2[i-1].find_all('td'))<=5:
                values=[tds[0].text.replace('\n', ''), tds[1].text.replace('\n', ''), tds[2].find('img').get('alt'), 
                    tds[3].text.replace('\n', ''), '', rows2[i-2].find_all('td')[5].text.replace('\n', '')]
                data=data.append(pd.Series(values, index=columns), ignore_index=True)
            else :
                values=[tds[0].text.replace('\n', ''), tds[1].text.replace('\n', ''), tds[2].find('img').get('alt'), 
                    tds[3].text.replace('\n', ''), '', rows2[i-1].find_all('td')[5].text.replace('\n', '')]
                data=data.append(pd.Series(values, index=columns), ignore_index=True)
        elif len(tds)<6:
            pass
        else :
            if tds[2].find('img')==None :
                values=[tds[0].text.replace('\n', ''), tds[1].text.replace('\n', ''), '', 
                        tds[3].text.replace('\n', ''), tds[4].text.replace('\n', ''), tds[5].text.replace('\n', '')]
                data =data.append(pd.Series(values, index=columns), ignore_index=True)
            else :
                values=[tds[0].text.replace('\n', ''), tds[1].text.replace('\n', ''), tds[2].find('img').get('alt'), 
                        tds[3].text.replace('\n', ''), tds[4].text.replace('\n', ''), tds[5].text.replace('\n', '')]
                data =data.append(pd.Series(values, index=columns), ignore_index=True)   
data
```
